### Histograms

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

ax1.plot(x, x);
ax2.scatter(x, x);
ax3.bar(x, x);
ax4.hist(x);
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0, 0].plot(x, x);
ax[0, 1].scatter(x, x);
ax[1, 0].bar(x, x);
ax[1, 1].hist(x);
```

# Plotting from Pandas DataFrames

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[$,.]", "", regex=True).str[:-2].astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot(kind="hist");
```

```python
heart_disease.plot(kind="hist", subplots=True, figsize=(10, 30));
```

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
fig, ax = plt.subplots()
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
```

```python
fig, ax = plt.subplots()

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(), linestyle="dashed")

scatter;
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(8, 8), sharex=True)

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(), linestyle="dashed")

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="dashed")

fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")

fig;
```

```python
x = np.random.randn(10, 4)
df = pd.DataFrame(x, columns=["a", "b", "c", "d"])
ax = df.plot(kind="bar")
ax;
```

```python
ax = df.plot(kind="bar")
ax.set(title="Random Number Bar Graph from DataFrame", xlabel="Row Number", ylabel="Random Number")
ax.legend().set_visible(True)
ax;
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots()

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(), linestyle="dashed")

scatter;
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(8, 8), sharex=True)

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
ax0.set_xlim([50, 80])
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(), linestyle="dashed")

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle="dashed")

fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")

fig;
```
